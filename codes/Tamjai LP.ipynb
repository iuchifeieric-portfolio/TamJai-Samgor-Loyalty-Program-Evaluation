```python
import pandas as pd

df = pd.read_csv("Tamjai_issues.csv")
df
```

```python
df['Problem Categories'] = df['Problem Categories'].astype(str)
df.info()
```

```python
import matplotlib.pyplot as plt

category_counts = df['Problem Categories'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('The Problem of Tamjai Samgor App')
plt.xlabel('Problem Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = [230, 40, 70, 30, 651]
ratings = ['5 Star', '4 stars', '3 stars', '2 stars', '1 stars']

plt.figure(figsize=(8, 5))
bars = plt.bar(ratings, counts, color='skyblue')
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Count')

# Add the counting numbers on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontsize=10)

plt.ylim(0, max(counts) * 1.1) # Adjust y-axis limit to make space for labels
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

counts = [230, 40, 70, 30, 651]
ratings = ['5 Star', '4 stars', '3 stars', '2 stars', '1 stars']

cmap = plt.colormaps.get_cmap('RdYlGn_r') 

colors = [cmap(i / (len(ratings) - 1)) for i in range(len(ratings))]

plt.figure(figsize=(8, 5))
bars = plt.bar(ratings, counts, color=colors) 
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Count')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontsize=10)

plt.ylim(0, max(counts) * 1.1)
plt.show()
```
