```python
# use the Google Play API

!pip install google-play-scraper
from google_play_scraper import app, reviews, Sort

result = app(
    'com.tamjaisamgor.hk',
    lang='en',
    country='hk'
)

reviews_result, continuation_token = reviews(
    'com.tamjaisamgor.hk',
    lang='en',
    country='hk',
    sort=Sort.NEWEST, # select the latest comments
    count=100
)
```
